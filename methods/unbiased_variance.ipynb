```python
import random
from matplotlib import pyplot as plt

import numpy as np
```

```python
def draw_sample(k=50):
    return random.choices(range(0,200), k=k)

def simulate_variance(samples=1000, ddof=0):
    observations = np.array([])
    var = np.array([])
    var_diff = np.array([])
    for i in range(0, samples):
        observations = np.append(observations, draw_sample())
        var = np.append(var, np.var(observations, ddof=ddof))
        if i > 0:
            var_diff = np.append(var_diff, var[i] - var[i-1])
    return {"var": var, "var_diff": var_diff}
        
```

 - *n* under estimates
 - *n-1* converges to the true variance
 - *n-2* over estimates the true variance

```python
n = simulate_variance()
n_1 = simulate_variance(ddof=1)
n_2 = simulate_variance(ddof=2)
```

```python
plt.plot(n["var"])
plt.plot(n_1["var"])
plt.plot(n_2["var"])
plt.legend(["n", "n - 1", "n - 2"])
```

```python
plt.plot(n["var_diff"])
plt.plot(n_1["var_diff"])
plt.plot(n_2["var_diff"])
plt.legend(["n", "n - 1", "n - 2"])
```
